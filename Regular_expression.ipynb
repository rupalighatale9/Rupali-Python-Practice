```python
import re

data = "Rupa is best girl is best is best is best think best in coding and sql"

find_text = r"best"

#match()
match = re.search(find_text,data)

if match:
    print("data find ...")
else:
    print("data not found...")


```

```python
# findall()

find_data = re.findall(find_text,data)
print(find_data)
print(len(find_data))
```
