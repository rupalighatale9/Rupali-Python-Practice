```python
import pandas as pd
print(pd.__version__)
```

```python
#pd.series()
data = ["Java","c++","Python","go","jsp","ruby","r"]

for i in range (len(data)):
    print(i,data[i])
```

```python
#using pandas series()
s = pd.Series(data)
print(s)
```

```python
#using index
indexx = [101,102,103,104,105,106,107]
s = pd.Series(data,index=indexx)
print(s)
```

```python
data = {
    'name':["Sara","Ashok","Shivam","Kajal","Aanchal","Ryan"],
    'age':[24,25,26,23,23,2],
    'city':["Mumbai","Pune","Goa","Jaipur","Udaipur","Surat"]
}
```

```python
indexx = [101,102,103,104,105,106]
#pd.DataFrame()
df = pd.DataFrame(data,index=indexx)
df
```

```python
#export file 
df.to_csv('data1.csv')
```

```python
#export file without index
df.to_csv('data01.csv',index=False)
```
