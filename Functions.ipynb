```python
# example -
def bill(price,gst,quantity):
    print('---------bill----------')
    print(f'price = {price}₹')
    print(f'gst = {gst}%')
    gst_amount = price*gst/100
    print(f'GST amount = {gst_amount}₹')
    total = price+gst_amount
    print(f'Total = {total}₹')
    print('----------------------')
    quantity_amount = price*quantity+gst_amount
    print(f'Quantity = {quantity}₹')
```

```python
bill(1200,18,10)
```
