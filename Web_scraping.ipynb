```python
import requests
from bs4 import BeautifulSoup

#url = "https://codeswithpankaj.com"
url = "https://www.amazon.in/ref=nav_logo"

res = requests.get(url)

soup = BeautifulSoup(res.text,"html.parser")
title = soup.title.text
print(title)
heading = soup.find_all("h1")

for i in heading:
    print(i.text)
```
