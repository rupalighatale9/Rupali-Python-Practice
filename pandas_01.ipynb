```python
import pandas as pd
print(pd.__version__)
```

```python
# pd.series()
data = ["java","c++","python","go","jsp","ruby","r"]

for i in range(len(data)):
    print(i,data[i])
```

```python
#using pandas series()
s = pd.Series(data)
print(s)
```

```python
# using index
indexx = [101,102,103,104,105,106,107]
s = pd.Series(data,index=indexx)
print(s)
```

```python
data = {
    "name":["rupali","disha","rahasya","shreya","akhilesh","prajwal"],
    "age":[22,23,24,25,26,27],
    "city":["mumbai","gujrat","banglore","pune","nandura","amravati"]
}
```
