```python
data ={
    'Name':'Rupali',
    'Age':22,
    'Height':5.6
}
```

```python
print(data)
```

```python
# Access data
print(data['Name'])
```

```python
# update data
data['Age']= 23
print(data)
```

```python
#add new data
data['city'] = 'Mumbai'
print(data)
```

```python
#keys() : Returns a list of all keys in the dictionary 
print(data.keys())

#values() : Returns a list of all values in the dictionary
print(data.values())

#items() : Returns a list of key value pairs of tuples
print(data.items())

#get(key,default): Returns the value for a key if it exists ,else returs a default value
print(data.get('Age'))
```

```python
#pop(key,default): removes and returns the value associated with the key
data.pop('Age')
```
