```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [10,20,30,40,50,60,70]
plt.plot(x,y)
plt.title("Line plot")
plt.xlabel("date")
plt.ylabel("sales")
plt.show()
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100) 
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,z)
plt.title("Line plot")
plt.xlabel("date")
plt.ylabel("sales")
plt.legend(["date",'cos'])
plt.show()

```

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x,y,color="pink",s=15,alpha=0.9)
plt.title("Line plot")
plt.xlabel("date")
plt.ylabel("sales")
plt.grid(True,alpha=0.9)
plt.show()
```

```python
# bar plot
categories = ["apple","banana","orange","grapes","watermelon"]
sales_25 = [25,40,32,20,60]
sales_26 = [45,65,76,89.90]

plt.bar(x=0.2,sales_25,color = ["red","yellow","orange","green","pink"],label="sales 2025",width=0.4)
plt.bar(x=0.2,sales_26,color = ["red","yellow","orange","green","pink"],label="sales 2026",width=0.4)

plt.title("Sales_of_Fruits")
plt.xlabel("sales")
plt.ylabel("product")
plt.xticks(x,categories)
plt.legend()
plt.grid(True,alpha=0.3)
plt.show()
```

```python
# bar plot
categories = ["Apple","Watermelon","Orange","Strawberry"]

sales_23 = [50,40,20,90]
sales_24 = [45,23,19,60]
sales_26 = [10,15,20,23]

x = np.arange(len(categories))

plt.bar(x-0.1,sales_23,label="sales 2023",width=0.4)
plt.bar(x-0.2,sales_24,label="sales 2024",width=0.4)
plt.bar(x-0.3,sales_26,label="sales 2026",width=0.4)
plt.title("Sales_of_Fruits")
plt.xticks(x,categories)
plt.xlabel("sales")
plt.ylabel("product")
plt.legend()
plt.grid(True,alpha=0.3)
plt.show()
```
