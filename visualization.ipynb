```python
import numpy as np
```

```python
x = np.linspace(0,10,100) 
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,z)
plt.title("Line plot")
plt.xlabel("date")
plt.ylabel("sales")
plt.legend(["date",'cos'])
plt.show()

```

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x,y,color="pink",s=15,alpha=0.9)
plt.title("Line plot")
plt.xlabel("date")
plt.ylabel("sales")
plt.grid(True,alpha=0.9)
plt.show()
```

```python
# bar plot
categories = ["apple","banana","orange","grapes","watermelon"]
sales_25 = [25,40,32,20,60]
sales_26 = [45,65,76,89.90]

plt.bar(x=0.2,sales_25,color = ["red","yellow","orange","green","pink"],label="sales 2025",width=0.4)
plt.bar(x=0.2,sales_26,color = ["red","yellow","orange","green","pink"],label="sales 2026",width=0.4)

plt.title("Sales_of_Fruits")
plt.xlabel("sales")
plt.ylabel("product")
plt.xticks(x,categories)
plt.legend()
plt.grid(True,alpha=0.3)
plt.show()
```

```python
# bar plot
categories = ["Apple","Watermelon","Orange","Strawberry"]

sales_23 = [50,40,20,90]
sales_24 = [45,23,19,60]
sales_26 = [10,15,20,23]

x = np.arange(len(categories))

plt.bar(x-0.1,sales_23,label="sales 2023",width=0.4)
plt.bar(x-0.2,sales_24,label="sales 2024",width=0.4)
plt.bar(x-0.3,sales_26,label="sales 2026",width=0.4)
plt.title("Sales_of_Fruits")
plt.xticks(x,categories)
plt.xlabel("sales")
plt.ylabel("product")
plt.legend()
plt.grid(True,alpha=0.3)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#set theme - white,dark,whitegrid,darkgrid,ticks
sns.set_theme(style='whitegrid',palette='pastel')
plt.rcParams['figure.figsize'] = (10,6) #default
```

```python
tips =sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
```

```python
# simple scatter plot two variable
#total bill
#tips

sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip'
)
plt.title("Total bill vs tips")
plt.show()
```

```python
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex', #color by gender
    style='size', #difference
    size='size',
    palette='deep'
)
plt.title("size vs tips")
plt.show()
```

```python
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex', #color by gender
    style='day', #difference
    size='size',
    palette='deep'
)
plt.title("Day vs tips")
plt.show()
```

```python
# line plot
flights = sns.load_dataset("flights")
flights.head()
```

```python
sns.lineplot(
    data=flights,
    x="year",
    y="passengers",
    hue="month"
)
plt.title("Airline data")
plt.show()
```

```python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex',
    col='time',
    row='day'
)
plt.show
```

```python
#histogram
sns.histplot(
    data=tips,
    x='total_bill',
    kde=True,
    bins=50
)
plt.title("Distribution of total bill")
plt.show()
```

```python
# category
sns.displot(
    data=tips,
    x='total_bill',
    kde=True,
    bins=40,
    hue='sex'

)
plt.title("Distribution of total bill")
plt.show()
```
