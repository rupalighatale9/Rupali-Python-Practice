```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [10,20,30,40,50,60,70]
plt.plot(x,y)
plt.title("Line plot")
plt.xlabel("date")
plt.ylabel("sales")
plt.show()
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100) 
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,z)
plt.title("Line plot")
plt.xlabel("date")
plt.ylabel("sales")
plt.legend(["date",'cos'])
plt.show()

```

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x,y,color="pink",s=15,alpha=0.9)
plt.title("Line plot")
plt.xlabel("date")
plt.ylabel("sales")
plt.grid(True,alpha=0.9)
plt.show()
```

```python
# bar plot
categories = ["apple","banana","orange","grapes","watermelon"]
sales = [25,40,32,20,60]

plt.bar(categories,sales,color = ["red","yellow","orange","green","pink"])
plt.grid(True,alpha=0.2)
plt.show()
```
