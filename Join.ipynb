```python
import pandas as pd


students = {
    'studentID': [101,102,103,104,105,108],
    'name':['ashok','Jaanvi','Shivam','Jamuna','lily','Abhi'],
    'Age': [24,22,24,26,21,24]
}

rseult = {
    'studentID': [101,102,103,104,105],
    'Subject': ['maths','java','c++','go','ruby'],
    'score': [88,75,90,55,86]
}
```

```python
students = pd.DataFrame(students)
students
```

```python
rseult = pd.DataFrame(rseult)
rseult
```

```python
left_merge = pd.merge(students,rseult,on='studentID',how='left')
left_merge
```

```python
right_merge = pd.merge(students,rseult,on='studentID',how='right')
right_merge
```

```python
outer_merge = pd.merge(students,rseult,on='studentID',how='outer')
outer_merge
```
